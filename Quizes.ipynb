Problem: Find the softmax values of a given matrix of scores.

```python
import tensorflow as tf
import numpy as np
import math
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def softmax(allScores):
    """ Compute the softmax value of a score.
    
    Parameters
    ----------
        allScores : ndarray
            nxm array with n = number of classes and m = number of samples
    Returns
    -------
        result : ndarray
            The calculated softmax values based on the formula: e^yi/sum(e^yi)
    """
    
    numClasses = allScores.shape[0]
    numSamples = allScores.shape[1]
    result = np.zeros_like(allScores)
    for i in xrange(numSamples):
        scores = allScores[:, i]
        expScores = np.array([math.exp(score) for score in scores])
        result[:, i] = [expScore/np.sum(expScores) for expScore in expScores]
    return result
```
